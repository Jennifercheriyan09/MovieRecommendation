```python
movies=movies.merge(credits,on="title")
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
import ast
```

```python
def convert(string):
    L=[]
    for i in ast.literal_eval(string):
        L.append(i["name"])
    return L
    
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies["keywords"]=movies["keywords"].apply(convert)
```

```python
def convert2(string):
    L=[]
    for i in ast.literal_eval(string):
        L.append(i["name"])
    return L
    
```

```python
movies["cast"]=movies["cast"].apply(convert2)
```

```python
movies["cast"]=movies["cast"].apply(lambda x:x[0:3])
```

```python
movies
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies
```

```python
def collapse(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies["tags"]=movies["overview"]+movies["genres"]+movies["keywords"]+movies["crew"]+movies["cast"]
```

```python
new_movies= movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

def recommend(movie_list):
    # Initialize a CountVectorizer to extract features from movie tags
    cv = CountVectorizer(max_features=5000, stop_words='english')
    
    # Combine the tags of all movies in the list into a single list
    combined_tags = []
    for movie in movie_list:
        index = new_movies[new_movies['title'] == movie].index[0]
        combined_tags.append(new_movies.iloc[index]['tags'])
    
    # Fit the CountVectorizer on combined tags and transform them into vectors
    vectorized_tags = cv.fit_transform(combined_tags)
    
    # Get feature names (tags) from the CountVectorizer
    feature_names = cv.get_feature_names_out()
    
    # Sum up the occurrences of each feature (tag)
    tag_frequencies = vectorized_tags.sum(axis=0).tolist()[0]
    
    # Create a dictionary to store tag frequencies
    tag_freq_dict = {}
    for tag, freq in zip(feature_names, tag_frequencies):
        tag_freq_dict[tag] = freq
    
    # Sort the tag frequencies in descending order
    sorted_tag_freq = sorted(tag_freq_dict.items(), key=lambda x: x[1], reverse=True)
    
    # Extract the top common attributes (excluding stop words)
    top_common_attributes = [tag for tag, freq in sorted_tag_freq if tag not in cv.stop_words_][:5000]

    # Find similar movies based on the top common attributes
    similar_movies = set()
    for tag in top_common_attributes:
        tag_index = new_movies[new_movies['tags'].str.contains(tag)].index
        similar_movies.update(tag_index)
    
    # Remove input movies from the set of similar movies
    similar_movies.difference_update(set(new_movies[new_movies['title'].isin(movie_list)].index))
    
    # Print recommended movies based on similarity
    print("Recommended movies:")
    for index in list(similar_movies)[:10]:
        print(new_movies.iloc[index]['title'])
```

```python
movie_list = ["The Dark Knight", "Inception", "Interstellar", "The Matrix", "The Shawshank Redemption",
              "Fight Club", "Pulp Fiction", "The Lord of the Rings: The Fellowship of the Ring"]
recommend(movie_list)
```
